```python
import torch, os
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_samples, silhouette_score, calinski_harabasz_score
from sklearn.manifold import TSNE
from joblib import dump, load
from ProjectDict import *

device = "cpu"
if torch.cuda.is_available():
    device = "cuda"
# device = 'dml' if torch.cuda.is_available() else 'cpu'
print(f"device: {device}")
print(torch.version.hip)
```

```python
rpath = os.path.join("inspected_data")
theme = laughing_man # laughing_man | wonder_boy
theme_en = laughing_man_en # laughing_man_en | wonder_boy_en
model_type = model_types.fine_tuned
df = pd.read_csv(os.path.join(rpath, f"{theme}_inspected.csv"),)
```

```python
#reference https://scikit-learn.org/stable/auto_examples/cluster/plot_dbscan.html
row, column = 2, 2
fig, axes = plt.subplots(row, column)
fig.set_size_inches(8, 8)
fig.suptitle(f"KMeans t-SNE")
fig.subplots_adjust(top=0.9)

fig2, axes2 = plt.subplots(row, column)
fig2.suptitle(f"KMeans Silhouette")
fig2.set_size_inches(8, 8)
# fig.suptitle(f"{model_type.capitalize()} KMeans t-SNE")

fig3, axes3 = plt.subplots(1,2)
fig3.set_size_inches(9, 4)
fig3.suptitle(f"Calinski-Harabasz Index")

# x = range(2, 2+row*column)
# x = range(2, 2+9)
x = [2, 4]
# not_skip = {2, 10}

for icol, model_type in enumerate([model_types.base, model_types.fine_tuned]):
    nested_labels = []
    silhouettes = []
    silhouettes_mean = []
    ch_indexes = []
    
    embeddings = torch.load(f"cache/{theme_en}_embeddings_{model_type}.pt")
    df_em = pd.DataFrame(embeddings.to("cpu"))
    
    folded_file_path = f"cache/{theme}_{model_type}_folded"
    if os.path.exists(folded_file_path):
        folded_np = np.load(folded_file_path)
    else:
        folded_np = TSNE(n_components=2).fit_transform(df_em)
        np.save(folded_file_path, folded_np)
    folded = torch.tensor(folded_np)
    
    for irow, k in enumerate(x):
        model_file_path = f"clustering/{theme}_{model_type}_KMeans_{k}.joblib"
        if os.path.exists(model_file_path):
            model = load(model_file_path)
        else:
            model = KMeans(n_clusters=k)
            model.fit(df_em)
            dump(model, model_file_path)
        labels = model.labels_

        nested_labels.append(model.labels_)
        silhouettes.append(silhouette_samples(df_em, labels))
        silhouettes_mean.append(silhouette_score(df_em, labels))
        
        ax = axes[irow][icol]

        unique_labels = set(range(k))

        colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

        for label, col in zip(unique_labels, colors):
            class_member_mask = labels == label

            xy = folded[class_member_mask].cpu()
            ax.plot(
                xy[:, 0],
                xy[:, 1],
                "o",
                markerfacecolor=tuple(col),
                markeredgecolor="k",
                markersize=3,
                markeredgewidth=0.1,
            )

        ax.set_title(f"{model_type.replace('_', '-').capitalize()} {k} clusters")

        # silhouettes.append(silhouette_samples(df_em, labels))
        # silhouettes_mean.append(silhouette_score(df_em, labels))
        print(f"Silhouette score: {silhouettes_mean[-1]:0.3f}")
        
        # labels = nested_labels[irow]
    
        n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise_ = list(labels).count(-1)

        ax2 = axes2[irow][icol]

        ax2.axvline(x=silhouettes_mean[irow], label=["Sihouette Score"])
        ax2.set_xlim(-0.1, 1)
        ax2.set_title(f"{model_type.replace('_', '-').capitalize()} #Clusters {k}, SS {silhouettes_mean[irow]:0.2f}")
        
        y_lower = 10
        for j in range(n_clusters_):
            jth_cluster_silhouette_values = silhouettes[irow][nested_labels[irow] == j]
            jth_cluster_silhouette_values.sort()

            y_upper = y_lower + jth_cluster_silhouette_values.shape[0]
            color = cm.nipy_spectral(float(j)/n_clusters_)
            ax2.fill_betweenx(np.arange(y_lower,y_upper),0,jth_cluster_silhouette_values,facecolor=color, edgecolor=color, alpha=0.7)

            y_lower = y_upper
        
    for irow, k in enumerate(range(2, 2+9)):
        model_file_path = f"clustering/{theme}_{model_type.replace('_', '-').capitalize()}_KMeans_{k}.joblib"
        if os.path.exists(model_file_path):
            model = load(model_file_path)
        else:
            model = KMeans(n_clusters=k)
            model.fit(df_em)
            dump(model, model_file_path)
            
        labels = model.labels_
        
        ch_indexes.append(calinski_harabasz_score(df_em, labels))
            
    ax3 = axes3[icol]
    ax3.plot(range(2,2+9), ch_indexes, '-o')
    ax3.set_xlabel("#Clusters")
    ax3.set_ylabel("Score")
    ax3.set_title(f"{model_type.replace('_', '-').capitalize()}")
    ax3.grid()

fig.tight_layout()

# plt.title(f"Estimated number of clusters: {n_clusters_}")
fig2.tight_layout()
fig2.supylabel("Index of Sample")

fig3.tight_layout()
plt.show()
    
```

```python
#reference https://scikit-learn.org/stable/auto_examples/cluster/plot_dbscan.html

row, column = 2, 2
fig, axes = plt.subplots(row, column)
fig.set_size_inches(8, 8)
fig.suptitle(f"DBSCAN t-SNE")
fig.tight_layout()
fig.subplots_adjust(top=.9)

# eps_range = np.linspace(0.1, 1.6, row * column)
eps_range = [0.2875, 0.475]

for icol, model_type in enumerate([model_types.base, model_types.fine_tuned]):
    nested_labels = []
    clusters = []
    silhouettes = []
    silhouettes_mean = []
    ch_indexes = []
    for irow, eps in enumerate(eps_range):
        
        model_file_path = f"clustering/{theme}_{model_type}_DBSCAN_{eps}.joblib"
        if os.path.exists(model_file_path):
            db = load(model_file_path)
        else:
            db = DBSCAN(eps=eps, min_samples=10) 
            db.fit(df_em)
            dump(db, model_file_path)
            
        labels = db.labels_
        nested_labels.append(labels)
        ax = axes[irow][icol]

        n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise_ = list(labels).count(-1)

        clusters.append(n_clusters_)
        if n_clusters_ == 1: continue

        print("Estimated number of clusters: %d" % n_clusters_, end = ", ")
        print("Estimated number of noise points: %d" % n_noise_, end = ", Silhouette Score: ")

        unique_labels = set(labels)
        core_samples_mask = np.zeros_like(labels, dtype=bool)
        core_samples_mask[db.core_sample_indices_] = True

        colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
        for k, col in zip(unique_labels, colors):
            if k == -1:
                # Black used for noise.
                col = [0, 0, 0, 0.1]

            class_member_mask = labels == k

            xy = folded[class_member_mask & core_samples_mask].cpu()
            ax.plot(
                xy[:, 0],
                xy[:, 1],
                "o",
                markerfacecolor=tuple(col),
                markeredgecolor="k",
                markersize=3,
                markeredgewidth=0.1,
            )

            xy = folded[class_member_mask & ~core_samples_mask].cpu()
            ax.plot(
                xy[:, 0],
                xy[:, 1],
                "o",
                markerfacecolor=tuple(col),
                markeredgecolor="k",
                markersize=3,
                markeredgewidth=0.1,
            )

        ax.set_title(f"{model_type.replace('_', '-').capitalize()}: eps {eps:0.2f}, {n_clusters_} clusters")

        silhouettes.append(silhouette_samples(df_em, labels))
        silhouettes_mean.append(silhouette_score(df_em, labels))
        ch_indexes.append(calinski_harabasz_score(df_em, labels))
        print(silhouette_score(df_em, labels))

# plt.title(f"Estimated number of clusters: {n_clusters_}")
fig.tight_layout()
plt.show()
```

```python
plt.plot(eps_range, clusters)
plt.title("#Clusters")
plt.xlabel("eps")
plt.grid()
plt.show()
```

```python
#reference https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html
fig, axes = plt.subplots(1, 2)
fig.suptitle(f"{model_type.replace('_', '-').capitalize()} DBSCAN Silhouette and Calinski-Harabasz Index")
fig.set_size_inches(8, 4)
fig.tight_layout()

# x_min, x_max = min(min(SS) for SS in silhouettes), max(max(SS) for SS in silhouettes)
x_min, x_max = -.3, 1

nested_labels = []
silhouettes = []
silhouettes_mean = []
ch_indexes = []

eps_range = np.linspace(0.1, 1.6, 9)
for i, eps in enumerate(eps_range):
    model_file_path = f"clustering/{theme}_{model_type}_DBSCAN_{eps}.joblib"
    if os.path.exists(model_file_path):
        db = load(model_file_path)
    else:
        db = DBSCAN(eps=eps, min_samples=10) 
        db.fit(df_em)
        dump(db, model_file_path)
        
    labels = db.labels_
    nested_labels.append(labels)
    
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise_ = list(labels).count(-1)
    
    silhouettes.append(silhouette_samples(df_em, labels))
    silhouettes_mean.append(silhouette_score(df_em, labels))
    ch_indexes.append(calinski_harabasz_score(df_em, labels))
    
    if i != 7: continue
    
    ax = axes[0]
    try:
        ax.set_title(f"Eps {eps:0.2f}, SS {silhouettes_mean[i]:0.2f}, Clusters {n_clusters_}")
    except IndexError:
        continue
    ax.set_xlim(x_min, x_max)
    
    ax.axvline(x=silhouettes_mean[i])
    y_lower = 10
    
    for j in range(n_clusters_):
        jth_cluster_silhouette_values = silhouettes[i][nested_labels[i] == j]
        jth_cluster_silhouette_values.sort()

        y_upper = y_lower + jth_cluster_silhouette_values.shape[0]
        color = cm.nipy_spectral(float(j)/n_clusters_)
        ax.fill_betweenx(np.arange(y_lower,y_upper),0,jth_cluster_silhouette_values,facecolor=color, edgecolor=color, alpha=0.7)

        y_lower = y_upper

axes[0].set_xlabel("Silhouette")
axes[0].set_ylabel("Sample")

eps_range = np.linspace(0.1, 1.6, 9)
ch_indexes = []
for irow, eps in enumerate(eps_range):
    model_file_path = f"clustering/{theme}_{model_type}_DBSCAN_{eps}.joblib"
    if os.path.exists(model_file_path):
        db = load(model_file_path)
    else:
        db = DBSCAN(eps=eps, min_samples=10) 
        db.fit(df_em)
        dump(db, model_file_path)
        
    labels = db.labels_
    ch_indexes.append(calinski_harabasz_score(df_em, labels))


axes[1].plot(eps_range[:len(ch_indexes)], ch_indexes, '-o')
axes[1].set_xlabel("eps")
axes[1].set_ylabel("Score")
axes[1].grid()
axes[1].set_title("Calinski-Harabasz Index")

fig.tight_layout()
plt.show()
```
